```python
from datetime import datetime

# oficios a utilizar
oficios_requeridos = ['Ingenieria', 'Desarrollo de software', 'Enfermeria', 'Docencia', 'Medicina', 'Chef', 'Encargado', 'Gerente', 'Auditor', 'Repartidor']

# Funcion para calcular la edad fertil y basarnos de ahi para sacar los puntos
def calcular_puntos_edad_fertil(fecha_nacimiento):
    fecha_actual = datetime.now()
    edad = fecha_actual.year - fecha_nacimiento.year - ((fecha_actual.month, fecha_actual.day) < (fecha_nacimiento.month, fecha_nacimiento.day))

    if 20 <= edad <= 35:
        return 15 - (edad - 20)
    else:
        return 0

# Función para obtener los datos de un progenitor
def obtener_datos_progenitor():
    sexo = input("Ingrese el sexo (M/F/N): ")
    fecha_nacimiento = datetime.strptime(input("Ingrese la fecha de nacimiento (YYYY-MM-DD): "), "%Y-%m-%d")
    grado_academico = input("Ingrese el grado académico (P/U/T/N): ")
    tiene_oficio = input("¿Tiene alguno de los 10 oficios requeridos? (S/N): ")

    puntos = calcular_puntos_edad_fertil(fecha_nacimiento)

    if sexo == 'F':
        puntos += 35  # Puntos adicionales por ser mujer en edad fértil

    if tiene_oficio == 'S' and oficio in oficios_requeridos:
        puntos += 8  # Puntos adicionales por tener un oficio requerido

    if grado_academico == 'P':
        puntos += 5  # Puntos adicionales por postgrado
    elif grado_academico == 'U':
        puntos += 3  # Puntos adicionales por título universitario

    return puntos

# Función principal
def main():
    # Preguntar si la familia es monoparental o biparental
    tipo_familia = input("¿La familia es monoparental (M) o biparental (B)? ")

    # Inicializar puntos totales
    puntos_totales = 0

    # Obtener datos del primer progenitor
    puntos_totales += obtener_datos_progenitor()

    # Si es una familia biparental, obtener datos del segundo progenitor
    if tipo_familia == 'B':
        puntos_totales += obtener_datos_progenitor()

    # Preguntar cuántos hijos hay en la familia
    num_hijos = int(input("¿Cuántos hijos tiene la familia? "))

    # Obtener datos de cada hijo y sumar los puntos
    for _ in range(num_hijos):
        puntos_totales += 8  # Puntos por cada hijo
        sexo_hijo = input("Ingrese el sexo del hijo (M/F/N): ")
        fecha_nacimiento_hijo = datetime.strptime(input("Ingrese la fecha de nacimiento del hijo (YYYY-MM-DD): "), "%Y-%m-%d")
        puntos_totales += calcular_puntos_edad_fertil(fecha_nacimiento_hijo)

    # Sumar puntos de los progenitores y mostrar el resultado
    puntos_totales += 6 if tipo_familia == 'M' else 12  # 6 puntos por familia monoparental, 12 por biparental
    print(f"El puntaje total de la familia es: {puntos_totales}")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```
